```python
import matplotlib.pyplot as plt
import numpy as np
from math import e

# 100 linearly spaced numbers
x = np.linspace(-2,2,100)
x2 = np.linspace(-10,10,100)

# the function, which is y = e^x here
y = np.maximum(0.0, x)
y2 = np.maximum(0.1*x, x)
y3 = 1 / (1 + e ** -x)
y4 = np.tanh(x)

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the function
plt.plot(x,y, 'y', label='ReLU', color="red")
plt.plot(x,y2, 'y', label='leaky ReLU', color="green")
#plt.plot(x2,y3, 'y', label='1/(1 + e^(-x))', color="blue")

#plt.plot(x,y4, 'y', label='sin(x)/cos(x)', color="orange")
plt.legend(loc='upper left')

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
  
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

x = np.linspace(-5, 5, 100)
z = 1/(1 + np.exp(-x))
  
plt.plot(x, z, 'y', label='Sigmoid', color="blue")
plt.plot(x, np.tanh(x), 'y', label='Tanh', color="orange")
plt.legend(loc='upper left')
plt.show()
```
